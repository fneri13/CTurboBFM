```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
L = 1
N = 120
```

```python
x = np.linspace(0, L, N)
X,Y = np.meshgrid(x,x)
```

```python
grid = {'X': X, 'Y': Y}
# Create output directory
os.makedirs("Grid", exist_ok=True)
with open("Grid" + '/grid_%02i_%02i.csv' %(N, N), 'w') as file:
    file.write(f"NDIMENSIONS=2\n")
    file.write(f"NI={N}\n")
    file.write(f"NJ={N}\n")
    file.write(f"NK=1\n")
    file.write("x,y,z\n")
    for i in range(N):
        for j in range(N):
            for k in range(1):
                file.write(f"{X[i,j]:.6f},{Y[i,j]:.6f},{0:.6f}\n")
```
