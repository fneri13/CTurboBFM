```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from Utils.styles import *
```

```python
input_filename = "Volume_CSV/results.csv"

with open(input_filename, 'r') as f:
        ni = int(f.readline().strip().split('=')[1])
        nj = int(f.readline().strip().split('=')[1])
        nk = int(f.readline().strip().split('=')[1])
    
    
df = pd.read_csv(input_filename, skiprows=3)
data = {col: df[col].to_numpy().reshape((ni, nj, nk)) for col in df.columns}
```

```python
zCheck = 30
ymax = data['y'][3, :, 3].max()
umax = data['Velocity Z'][3, :, 3].max()
tmax = data['Temperature'][3, :, 3].max()
tmin = data['Temperature'][3, :, 3].min()
smax = data['Entropy'][3, :, 3].max()

plt.figure()
plt.plot(data['Velocity Z'][3, :, 3]/umax, data['y'][3, :, 3]/ymax, '-',label='Current')
plt.legend()
plt.ylabel(r'$y/H$ [-]')
plt.xlabel(r'$u/u_{max}$ [-]')
plt.grid(alpha=0.2)
plt.savefig('Pictures/couette_uz.pdf', bbox_inches='tight')

plt.figure()
plt.plot((data['Temperature'][3, :, 3])/tmax, data['y'][3, :, 3]/ymax, '-',label='Current')
plt.legend()
plt.ylabel(r'$y/H$ [-]')
plt.xlabel(r'$T/T_{max}$ [-]')
plt.grid(alpha=0.2)
plt.savefig('Pictures/couette_t.pdf', bbox_inches='tight')

plt.figure()
plt.plot((data['Entropy'][3, :, 3])/smax, data['y'][3, :, 3]/ymax, '-',label='Current')
plt.legend()
plt.ylabel(r'$y/H$ [-]')
plt.xlabel(r'$s/s_{max}$ [-]')
plt.grid(alpha=0.2)
plt.savefig('Pictures/couette_s.pdf', bbox_inches='tight')

```
